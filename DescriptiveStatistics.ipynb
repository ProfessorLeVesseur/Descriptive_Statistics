#**Descriptive Statistics** 

Refers to a set of techniques for summarizing and displaying data. Although in most cases the primary research question will be about one or more statistical relationships between variables, it is also important to describe each variable individually. For this reason, we begin by looking at some of the most common techniques for describing single variables.

Descriptive statistics are used to summarize and describe the main features of the data. The following are some common ways to report descriptive statistics:



*   **Measures of central tendency**: You can report the mean, median, and mode of the data.
*   **Measures of variability**: You can report the range, interquartile range (IQR), variance, and standard deviation of the data.
*   **Skewness and Kurtosis**: You can report the skewness and kurtosis values to describe the shape of the data.
*   **Frequency distributions**: You can report frequency distributions

##**Frequency Distribution**

In a frequency distribution, the data is grouped into classes or intervals, and the frequency of each class is counted and displayed. This provides a visual representation of the distribution of the data, showing how the values are spread out and how often certain values occur. Frequency distributions can be represented in various forms such as a histogram, bar graph, or table.

####**Examples of populations with different types of distributions include:**

**Negatively skewed** (left-skewed) distributions: These are distributions in which the values are concentrated on the right side of the scale and the tail extends to the left. Examples of populations with negatively skewed distributions include:
* **Income**: In many countries, a few individuals have very high incomes while the majority have relatively low incomes. This can result in a negatively skewed distribution of income.
* **Lifespans**: The lifespan of humans follows a negatively skewed distribution, with the majority of people living to an average age, but a small proportion living to much older ages.

**Symmetrical distributions**: These are distributions in which the values are spread evenly on either side of the center, so that the mean, median, and mode are all equal. Examples of populations with symmetrical distributions include:
* **Personality traits**: Within a population, the distribution of certain personality traits, such as agreeableness or neuroticism, may be symmetrical, with the majority of individuals having scores near the average and fewer individuals having scores far from the average.
* **Intelligence quotient** (IQ) scores: Within a population, the distribution of IQ scores may be symmetrical, with the majority of individuals having scores near the average and fewer individuals having scores far from the average.
* **Memory recall**: Within a population, the distribution of scores on a memory recall task may be symmetrical, with the majority of individuals having scores near the average and fewer individuals having scores far from the average.
* **Heights** of individuals within a population: Within a population, the height of individuals may be distributed evenly around the average height, resulting in a symmetrical distribution.

**Positively skewed** (right-skewed) distributions: These are distributions in which the values are concentrated on the left side of the scale and the tail extends to the right. Examples of populations with positively skewed distributions include:
* **Test scores**: In a school test, many students may get average or lower scores, while a few students score very high. This can result in a positively skewed distribution of test scores.

It's important to note that real-world populations often have complex distributions that cannot be characterized as being purely symmetrical, positively skewed or negatively skewed, but rather have some degree of skewness in one direction or another.

##**Skewness**

The data is said to be skewed when its probability distribution is not symmetric around the mean of that data. Three main scenarios can happen depending on the value of the skewness.

```python
# Use the seed to get the same results for randomization
np.random.seed(2023)

def plot_skewness(data, label):
  plt.hist(data, density=True, color='midnightblue', alpha=0.7, label=label)
  plt.axvline(data.mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
  plt.axvline(np.median(data), color='yellow', linestyle='dashed', linewidth=2, label='Median')
  plt.legend()
  plt.show()
```

###**Symmetric/Normal Distribution**

The skewness is symmetric when the data follows a normal distribution. In this case Mean = Median.

```python
# Normal distribution
normal_data = np.random.normal(0, 1, 1000)
label = 'Normal: Symetric Skewness'
plot_skewness(normal_data, label)
```

###**Positive Skewness**

There is a **positive skewness** or right skewness when its peak toward the lower end of its range and a relatively long positive tail. This means that the right side of the mean value contains more value, and the mean is to the right side of the median. In this case, we have Mean > Median.


```python
# Exponential distribution
exp_data = np.random.exponential(1, 1000)
label = 'Exponential: Positive Skewness'
plot_skewness(exp_data, label)
```

###**Negative Skewness**

There is a **negative skewness** or lest skewness when its peak toward the lohigherwer end of its range and a relatively long negative tail. This means that the left side of the mean value contains more value, and the mean is to the left side of the median. In this case, we have Mean < Median.

```python
# Beta 
beta_data = beta.rvs(5, 2, size=10000)
label = 'Beta: Negative Skewness'
plot_skewness(beta_data, label)
```

##**Kurtosis**

Statisticians now understand that kurtosis is a measure of **tailedness**, not *peakedness*.

The kurtosis metric quantifies the proportion of the distribution’s weight in its tails compared to the rest of the distribution. It tells us if the data is **spread out or concentrated around the mean**.

A distribution with a higher concentration around the mean is said to have **high kurtosis**. A **low kurtosis** is related to a more flat distribution with fewer data concentrated around the mean.

Kurtosis is used to check whether the data follows a **normal distribution**, and also for detecting the presence of **outliers** in the data.

Screenshot 2023-02-05 at 7.28.36 PM.png

```python
x = np.linspace(-5, 5, 100)
ax = plt.subplot()
distnames = ['laplace', 'norm', 'uniform']
# distnames = ['Leptokurtic', 'Normal Distribution', 'Platykurtic']

for distname in distnames:
    if distname == 'uniform':
        dist = getattr(stats, distname)(loc=-2, scale=4)
    else:
        dist = getattr(stats, distname)
    data = dist.rvs(size=1000)
    kur = kurtosis(data, fisher=True)
    y = dist.pdf(x)
    ax.plot(x, y, label="{}, {}".format(distname, round(kur, 3)))
    ax.legend()
```

##What is the difference between skewness and kurtosis?

Skewness and kurtosis are both important measures of a distribution’s shape.

*   **Skewness** measures the asymmetry of a distribution.
*   **Kurtosis** measures the heaviness of a distribution’s tails relative to a normal distribution.

###**What is the difference between Kurtosis and Excess Kurtosis?**
Kurtosis and excess kurtosis are measures of the tailedness of a distribution.

Kurtosis is a measure of the degree to which a distribution is peaked or flat compared to a normal distribution. A normal distribution has a kurtosis of 3. A distribution with a positive kurtosis is said to be leptokurtic, meaning it has a higher peak and fatter tails than the normal distribution. A distribution with a negative kurtosis is said to be platykurtic, meaning it has a flatter peak and thinner tails than the normal distribution.

Excess kurtosis, on the other hand, is the difference between the kurtosis of a distribution and the kurtosis of the normal distribution. The normal distribution has a kurtosis of 0, so its excess kurtosis is also 0. Excess kurtosis is used to compare the peakedness of a distribution relative to the normal distribution. If a distribution has an excess kurtosis of 0, it means it is mesokurtic and has the same level of peakedness as the normal distribution. If a distribution has an excess kurtosis greater than 0, it means it is leptokurtic and has a higher peak than the normal distribution. If a distribution has an excess kurtosis less than 0, it means it is platykurtic and has a flatter peak than the normal distribution.

##**Cut-Off Values for Excess Kurtosis**

The cut-off values for kurtosis depend on the type of distribution and the level of statistical significance. Kurtosis is a measure of the peakedness or flatness of a distribution compared to a normal distribution. 

The **normal/mesokurtic distribution** has a kurtosis value of between -1 and 1, meaning the distribution is similar to a normal distribution.

The **platykurtic/uniform distribution** has a kurtosis value less than -1, meaning the distribution is flatter than a normal distribution. 

The **leptokurtic distribution** has a kurtosis value greater than 1, meaning the distribution is more peaked than a normal distribution. 

However, it's important to note that these cut-off values are general guidelines and may vary based on the specific analysis and level of significance. In some cases, different values may be used as cut-offs, and the interpretation of kurtosis may also vary depending on the research question and the domain of study.

**Excel KURT Function**:

The Excel KURT function calculates the excess kurtosis, which is a measure of the tailedness of a distribution relative to the normal distribution. The excess kurtosis is calculated as the kurtosis of the data minus the kurtosis of the normal distribution, which is 3.

In other words, the KURT function calculates the fourth standardized moment about the mean and adjusts for the fact that the normal distribution has a kurtosis of 3. The result of the KURT function is expressed in units of standard deviations from the normal distribution, so a positive result indicates a leptokurtic distribution (more peaked than normal) and a negative result indicates a platykurtic distribution (flatter than normal).

Distribution on the basis of skewness value:

* Skewness > 0: **Positive**/right Skew, more weight in the left tail of the distribution. In other words, there are more values on the lower end of the data range and fewer on the higher end. 
* Skewness < 0: **Negative**/left Skew, more weight in the right tail of the distribution. In other words, there are more values on the higher end of the data range and fewer on the lower end. 

**Kurtosis** is a statistical term and an important characteristic of frequency distribution. It determines whether a distribution is heavy-tailed in respect of the normal distribution. It provides information about the shape of a frequency distribution.

* Kurtosis = 3: **Normal** distribution
* Kurtosis < 3: **Playkurtic** distribution
* Kurtosis > 3: **Leptokurtic** distribution and it signifies that it produces more outliers than the normal distribution
